### exp

```python
np.exp(1)
```

```python
x = np.arange(-1, 3)
y = np.exp(x)
print(y)
```

```python
import matplotlib.pyplot as plt
x = np.linspace(-2, 2, 20)
y = np.exp(x)
plt.plot(x, y)
plt.show()
```

### $$sin, cos$$

```python
print(np.sin(0), np.sin(pi/3), np.sqrt(3)/2)
```

```python
x = np.linspace(0, 2 * pi, 50)
plt.plot(x, np.sin(x), label='sin(x)')
plt.plot(x, np.cos(x), label='cos(x)')

plt.legend()
plt.show()
```

### $$ tan $$

```python
x = np.linspace(-99 * pi / 200, pi* 99 / 200, 1000)

plt.plot(x, np.tan(x), label="tan")

plt.legend()
plt.show()
```

### log

```python
x = np.linspace(0.1, 7, 100)

plt.plot(x, np.log(x), label="tan")

plt.legend()
plt.show()
```
