# Random Sampling

- random.rand
    - 결과:  $[0,1)$
- random.randn
    - 결과:  평균 0인 정규 분포
- random.ranf / random.random / random.sample / random.random_sample
    - 결과:  $[0,1)$
- random.randint
    - 결과:  범위 지정
- random.uniform
    - 결과:  범위 지정
- random.normal
    - 결과:  평균, 분산 지정

```python
import numpy as np
```

### random.rand
$[0,1)$

```python
print(np.random.rand())
```

```python
print(np.random.rand(2, 5))
```

### random.randn

 “standard normal” distribution: 정규 분포를 따름

```python
print(np.random.randn())
```

```python
print(np.random.randn(6))
```

```python
import matplotlib.pyplot as plt
x = np.random.randn(100000) * 100
x = np.array(x, np.int16) / 100
unique, counts = np.unique(x, return_counts=True)
plt.plot(unique, counts)
plt.show()
```

### random.ranf / random.random / random.sample / random.random_sample

random.ranf == random.random == random.sample == random.random_sample<br/>
파라미터: int or tuple<br/>
결과: $[0, 1)$

```python
a = np.random.ranf((3, 5))
print(a)
```

### random.randint

```python
np.random.randint(100)
```

```python
print(np.random.randint(4, size=(2,4)))
```

```python
print(np.random.randint(3, 5, size=(2,4)))
```

### random.uniform

```python
print(np.random.uniform())
```

```python
print(np.random.uniform(5, 9, size= (3, 5)))
```

```python
import matplotlib.pyplot as plt

x = range(1000)
y = np.random.uniform(5, 9, size=1000)

plt.scatter(x, y, s=1)
plt.show()
```

### random.normal

```python
print(np.random.normal())
```

```python
print(np.random.normal(100, 10))
```

```python
print(np.random.normal(100, 10, size = (3, 5)))
```

```python
import matplotlib.pyplot as plt

x = range(10000)
y = np.random.normal(100, 10, size= 10000)

plt.scatter(x, y, s=1)
plt.show()
```
